```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

deva=pd.read_csv('energy.csv')
print(deva)
srh=deva[deva["State"]=="Andhra"]
rcb=deva[deva["State"]=='Karnataka']
# line chart andhra karnataka
fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'],color='blue')
diag[1].plot(rcb['Date'],rcb['Consumption'],color='red')

#bar chart Telengana Tripura
revanth=deva[deva['State']=='Telengana']
cm=deva[deva['State']=='Tripura']

fig,diag=plt.subplots(2)
diag[0].bar(revanth['Date'],revanth['Consumption'],color='blue')
diag[1].bar(cm['Date'],cm['Consumption'],color='red')

#pie chart
state1='Delhi'
state2='Odisha'
state3='Up'
state4='Sikkim'
cse5=[state1,state2]

cse6=deva[deva['State']==state1]['Consumption'].sum()
cse4=deva[deva['State']==state2]['Consumption'].sum()
cse3=[cse6,cse4]

fig,diag=plt.subplots(2)
diag[0].pie(cse3,labels=cse5,autopct='%1.1f%%')

cse5=[state3,state4]
cse6=deva[deva['State']==state3]['Consumption'].sum()
cse4=deva[deva['State']==state4]['Consumption'].sum()
cse3=[cse6,cse4]
diag[1].pie(cse3,labels=cse5,autopct='%1.1f%%')
```
